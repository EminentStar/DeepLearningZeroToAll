```python
# Lab 2 Linear Regression
import tensorflow as tf
```

```python
x_train = [1, 2, 3]
y_train = [1, 2, 3]

X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])
# Variable이란 개념은 기존 프로그램의 변수와는 다름.. 텐서플로가 사용하는 variable,,
# 이 이야기는 텐서플로우를 실행시키면 자체적으로 변경시키는 값, 혹은 다른말론 트레이너블한 variable이다.
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
# * Our hypothesis XW+b
hypothesis = X * W + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
"""
t = [1., 2., 3., 4.]
tf.reduce_mean(t) # 2.5

"""
# 걍 현재 magic이라고 생각해라.
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost) # 이게 뭐 그래프상의 노드가 되겟지
```

```python
sess = tf.Session()

sess.run(tf.global_variables_initializer())
"""
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))

"""
for step in range(2001):
    cost_val, W_val, b_val, _ = \
        sess.run([cost, W, b, train],
            feed_dict={X:[1,2,3], Y:[1, 2, 3]}
            )
    if step % 20 == 0:
        print(step, cost_val, W_val, b_val)

```

```python
# Test model
print(sess.run(hypothesis, feed_dict={X: [5.]}))
print(sess.run(hypothesis, feed_dict={X: [2.5]}))
print(sess.run(hypothesis, feed_dict={X: [1.5, 3.5]}))
```

```python
# 필요한 값들을 placeholder로 넘겨줄 수 있음.
a = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)
adder_node = a + b

print(sess.run(adder_node, feed_dict={a:3, b: 4.5}))
print(sess.run(adder_node, feed_dict={a:[1, 3], b: [2, 4]}))
```
